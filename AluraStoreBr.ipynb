### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
#calculando o faturamento das lojas
faturamento_loja1 = loja['Preço'].astype(float).sum()
faturamento_loja2 = loja2['Preço'].astype(float).sum()
faturamento_loja3 = loja3['Preço'].astype(float).sum()
faturamento_loja4 = loja4['Preço'].astype(float).sum()

#mostrando os faturamentos
print("Faturamento total de cada loja:")
print(f"Faturamento loja 1: R${faturamento_loja1:.2f}")
print(f"Faturamento loja 2: R${faturamento_loja2:.2f}")
print(f"Faturamento loja 3: R${faturamento_loja3:.2f}")
print(f"Faturamento loja 4: R${faturamento_loja4:.2f}")
```

# 2. Vendas por Categoria


```python
#calculando a quantidade de vendas por categoria de cada loja
def vendas_por_categoria(loja, numero_loja):
    #agrupando os dados por categoria e somando as vendas
    vendas_categoria = loja.groupby('Categoria do Produto').size() #contando o número de vendas por categoria
    #ordenando os dados
    vendas_categoria = vendas_categoria.sort_values(ascending=False)
    return vendas_categoria

print(f"\nQuantidade de vendas por categoria da loja 1:")
print(vendas_por_categoria(loja, 1))
print(f"\nQuantidade de vendas por categoria da loja 2:")
print(vendas_por_categoria(loja2, 2))
print(f"\nQuantidade de vendas por categoria da loja 3:")
print(vendas_por_categoria(loja3, 3))
print(f"\nQuantidade de vendas por categoria da loja 4:")
print(vendas_por_categoria(loja4, 4))
```

# 3. Média de Avaliação das Lojas

```python
#calculculando a media de avaliação dos produtos de cada loja
def media_avaliacao(loja, numero_loja):
    media = loja['Avaliação da compra'].mean()
    return media

print(f"\nMédia de avaliação dos produtos da loja 1: {media_avaliacao(loja, 1):.2f}")
print(f"Média de avaliação dos produtos da loja 2: {media_avaliacao(loja2, 2):.2f}")
print(f"Média de avaliação dos produtos da loja 3: {media_avaliacao(loja3, 3):.2f}")
print(f"Média de avaliação dos produtos da loja 4: {media_avaliacao(loja4, 4):.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
#analisado quais podutos foram mais e menos vendidos

def produtos_mais_vendidos(loja, numero_loja):
    produtos_vendidos = loja.groupby(['Produto', 'Categoria do Produto']).size()
    produtos_vendidos = produtos_vendidos.sort_values(ascending=False)
    return produtos_vendidos.head(3)

def produtos_menos_vendidos(loja, numero_loja):
    produtos_vendidos = loja.groupby(['Produto', 'Categoria do Produto']).size()
    produtos_vendidos = produtos_vendidos.sort_values(ascending=True)
    return produtos_vendidos.tail(3)

print(f"\nProdutos mais vendidos da loja 1:")
print(produtos_mais_vendidos(loja, 1))
print(f"\nProdutos menos vendidos da loja 1:")
print(produtos_menos_vendidos(loja, 1))
print(f"\nProdutos mais vendidos da loja 2:")
print(produtos_mais_vendidos(loja2, 2))
print(f"\nProdutos menos vendidos da loja 2:")
print(produtos_menos_vendidos(loja2, 2))
print(f"\nProdutos mais vendidos da loja 3:")
print(produtos_mais_vendidos(loja3, 3))
print(f"\nProdutos menos vendidos da loja 3:")
print(produtos_menos_vendidos(loja3, 3))
print(f"\nProdutos mais vendidos da loja 4:")
print(produtos_mais_vendidos(loja4, 4))
print(f"\nProdutos menos vendidos da loja 4:")
print(produtos_menos_vendidos(loja4, 4))
```

# 5. Frete Médio por Loja


```python
#calculando o custo de frete médio de cada loja
def custo_frete_medio(loja, numero_loja):
    frete_medio = loja['Frete'].mean()
    return frete_medio

print(f"\nCusto de frete médio da loja 1: R${custo_frete_medio(loja, 1):.2f}")
print(f"Custo de frete médio da loja 2: R${custo_frete_medio(loja2, 2):.2f}")
print(f"Custo de frete médio da loja 3: R${custo_frete_medio(loja3, 3):.2f}")
print(f"Custo de frete médio da loja 4: R${custo_frete_medio(loja4, 4):.2f}")
```

# Gráficos

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter
```

```python
###################Grafico do faturamento total de cada loja

def criar_grafico_faturamento(faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4):
    # Função para formatar os valores em reais
    def formato_reais(x, p):
        return f'R${x:,.2f}'.replace(',', '.')
    
    # Criando o gráfico
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'], 
           [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4])
    
    # Mostrar os valores em reais
    ax.yaxis.set_major_formatter(FuncFormatter(formato_reais))
    

    plt.title('Faturamento total de cada loja')
    plt.xlabel('Lojas')
    plt.ylabel('Faturamento')
    plt.tight_layout()
    
    import os
    os.makedirs('graficos', exist_ok=True)
    

    plt.savefig('graficos/faturamento_lojas.png')
    plt.show()


criar_grafico_faturamento(faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4)



```

```python
###################Grafico das vendas por categoria de cada loja

def criar_grafico_categorias():

    vendas_loja1 = vendas_por_categoria(loja, 1)
    vendas_loja2 = vendas_por_categoria(loja2, 2)
    vendas_loja3 = vendas_por_categoria(loja3, 3)
    vendas_loja4 = vendas_por_categoria(loja4, 4)
    
    todas_categorias = pd.concat([
        vendas_loja1, vendas_loja2, vendas_loja3, vendas_loja4
    ]).index.unique()
    
    fig, ax = plt.subplots(figsize=(15, 8))
    

    x = np.arange(len(todas_categorias))
    width = 0.2
    

    barra1 = ax.bar(x - width*1.5, [vendas_loja1.get(cat, 0) for cat in todas_categorias], width, label='Loja 1', color='blue')
    barra2 = ax.bar(x - width/2, [vendas_loja2.get(cat, 0) for cat in todas_categorias], width, label='Loja 2', color='#12dba8')
    barra3 = ax.bar(x + width/2, [vendas_loja3.get(cat, 0) for cat in todas_categorias], width, label='Loja 3', color='#12db34')
    barra4 = ax.bar(x + width*1.5, [vendas_loja4.get(cat, 0) for cat in todas_categorias], width, label='Loja 4', color='#af12db')
    

    ax.set_ylabel('Quantidade de Produtos Vendidos')
    ax.set_title('Vendas por Categoria e por Loja')
    ax.set_xticks(x)
    ax.set_xticklabels(todas_categorias, rotation=45, ha='right')
    ax.legend()
    
    # funcao que adiciona percentual
    def adicionaPorcentagem(barra, vendas):
        for index, bar in enumerate(barra):
            height = bar.get_height()
            total_categoria = sum(v.get(todas_categorias[index], 0) for v in [vendas_loja1, vendas_loja2, vendas_loja3, vendas_loja4])
            if total_categoria > 0:
                porcentagem = height/total_categoria*100
                ax.text(bar.get_x() + bar.get_width()/2., height,
                       f'{porcentagem:.1f}%',
                       ha='center', va='bottom')
    
    adicionaPorcentagem(barra1, vendas_loja1)
    adicionaPorcentagem(barra2, vendas_loja2)
    adicionaPorcentagem(barra3, vendas_loja3)
    adicionaPorcentagem(barra4, vendas_loja4)
    

    plt.tight_layout()
    
    plt.savefig('graficos/vendas_por_categoria_e_loja.png')
    plt.show()
    


# Criando o gráfico
criar_grafico_categorias()
```

```python
###################Grafico da média de avaliação dos produtos

def criar_grafico_avaliacao():

    media_loja1 = media_avaliacao(loja, 1)
    media_loja2 = media_avaliacao(loja2, 2)
    media_loja3 = media_avaliacao(loja3, 3)
    media_loja4 = media_avaliacao(loja4, 4)

    medias = [media_loja1, media_loja2, media_loja3, media_loja4]
    lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
    cores = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']
    

    fig, ax = plt.subplots(figsize=(10, 6))
    

    barras = ax.barh(lojas, medias, color=cores)
    

    ax.set_title('Média de Avaliação por Loja')
    ax.set_xlabel('Média de Avaliação')
    ax.set_ylabel('Lojas')
    

    for i, v in enumerate(medias):
        ax.text(v, i, f' {v:.2f}', va='center')
    

    plt.tight_layout()
    

    plt.savefig('graficos/avaliacoes_lojas_barh.png')
    plt.show()
    


# Criando o gráfico
criar_grafico_avaliacao()
```

```python
###################Grafico dos produtos mais vendiddos de cada lokja
def grafico_produtos_mais_vendidos():
    fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2,figsize=(15, 12))

    produtos_loja1 = produtos_mais_vendidos(loja, 1)
    produtos_loja2 = produtos_mais_vendidos(loja2, 2)
    produtos_loja3 = produtos_mais_vendidos(loja3, 3)
    produtos_loja4 = produtos_mais_vendidos(loja4, 4)

    cores = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']

    def criar_subplot_pie(ax, dados, titulo):
        wedges, texts, autotexts = ax.pie(dados.values,
                                          labels = [prod[0] for prod in dados.index],
                                          colors = cores,
                                          autopct='%1.1f%%')
        ax.set_title(titulo)

        
        plt.setp(autotexts, size=8, weight="bold")

    criar_subplot_pie(ax1, produtos_loja1, 'Produtos Mais Vendidos - Loja 1')
    criar_subplot_pie(ax2, produtos_loja2, 'Produtos Mais Vendidos - Loja 2')
    criar_subplot_pie(ax3, produtos_loja3, 'Produtos Mais Vendidos - Loja 3')
    criar_subplot_pie(ax4, produtos_loja4, 'Produtos Mais Vendidos - Loja 4')

    plt.tight_layout()

    plt.savefig('graficos/produtos_mais_vendidos.png')
    plt.show()


grafico_produtos_mais_vendidos()
```

```python
###################Grafico dos fretes médios de cada loja

def grafico_fretes():
    frete_loja1 = custo_frete_medio(loja, 1)
    frete_loja2 = custo_frete_medio(loja2, 2)
    frete_loja3 = custo_frete_medio(loja3, 3)
    frete_loja4 = custo_frete_medio(loja4, 4)

    fretes = [frete_loja1, frete_loja2, frete_loja3, frete_loja4]
    lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
    cores = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']

    fig, ax = plt.subplots(figsize=(10, 6))

    barras = ax.bar(lojas, fretes, color=cores)

    ax.set_title('Custo de Frete Médio por Loja')
    ax.set_xlabel('Lojas')
    ax.set_ylabel('Custo de Frete Médio (R$)')

    def formatar_reais(x, pos):
        return f'R${x:,.2f}'.replace(',', '.')
    
    ax.yaxis.set_major_formatter(FuncFormatter(formatar_reais))

    for i, v in enumerate(fretes):
        ax.text(i, v, formatar_reais(v, None),
                ha='center', va='bottom')
        
    plt.tight_layout()
    plt.savefig('graficos/fretes_medio.png')
    plt.show()

grafico_fretes()
```
